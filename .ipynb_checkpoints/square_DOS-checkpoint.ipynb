```python
import matplotlib.pyplot as plt
import numpy as np
import tangent_boundaries as tb
#import sys
from math import pi

sigma_0 = np.array([[1,0],[0,1]])
sigma_x = np.array([[0,1],[1,0]])
sigma_y = np.array([[0,-1j],[1j,0]])
sigma_z = np.array([[1,0],[0,-1]])

Nx = 31 #Assume translational invariance in x direction
Ny = 31
nbands = int((Nx*Ny)/100*16)
print('Bands calculated: ',nbands) 

#i = int(sys.argv[1])
#ntasks = 20
i =  0
ntasks = 1

parameters = dict(
    Nx = Nx,
    Ny = Ny,
    B1 = 0, # no magnetic field
    N1 = 0, #
    d1 = 0, # These are irrelevant for B1 = 0
    N2 = 0, #
    potential = lambda x,y:0.0*np.random.rand(Ny,Nx),
    mass = lambda x,y:0*x,
    disorder = 0,
    theta = -(pi/2)*(i/ntasks),
)


spectrum = tb.get_spectrum(parameters, number_of_bands = nbands)
```

```python
ran = max(spectrum)

histogram, bin_edges = np.histogram(spectrum, bins = 41, range = (-ran,ran), density = False )

y = histogram
x = (bin_edges[1:]+bin_edges[:-1])/2
dx = (bin_edges[1]-bin_edges[0])/2

print(max(histogram))

fig = plt.figure(figsize = (7,7))
ax = fig.add_subplot(111)


ax.bar(x,y,width = 2*dx)

ax.set_ylabel(r"$\rho$ $units$",fontsize = 18)
ax.set_xlabel(r"$E$ $units$",fontsize = 18)
ax.set_ylim(0,2*(Nx+Ny))

fig.show()
```
